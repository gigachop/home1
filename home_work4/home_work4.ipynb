```python
documents = [ {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, 
             {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
             {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} 
            ]

directories = { '1': ['2207 876234', '11-2'], '2': ['10006'], '3': [] }
```

```python
# ЭТО ОСНОВНОЙ КОД

# ПУНКТ 1


def pasport_name(pasport):
    """
    эта функция берёт из файла documents - номер и имя. если вводимый пользователем номер будет совподать с номером из documents
    то функция вернёт имя, а если не найдёт то выведет что его нету
    """
    
    for document in documents: # перебираю все документы
            
        num_pasport = document['number'] # сюда номер документа
        name = document['name'] # сюда имя
        
        if pasport == num_pasport: # если документ найден то тогда вывожу имя            
            return name
        
        else:
            continue
        
    netu = 'Документ не найден в базе'
    return netu

def main_pasport_name():
    """
    эта функция уже будет выводить результат об имени
    """
    
    pasport = input('Введите номер документа: ') # сдесь вводится номер документа
        
    if pasport_name(pasport) != 'Документ не найден в базе':        
        print('Владелец документа:', pasport_name(pasport))
    
    else:
        print(pasport_name(pasport))


# ПУНКТ 2      
def shelf_pasport(pasport):
    """
    эта функция берёт из файла directories номер полки и документы в ней, далее она проверяет есть ли вводимый
    пользоватилем документ на полке, если есть возвращяет номер полки. если его вообще нет то возвращает что его нет
    """
    for shelf,document in directories.items():
        
        if pasport in document:
            return shelf
        else:
            continue
    netu = 'Документ не найден в базе'
    return netu

def main_shelf_pasport():
    """
    эта функция уже будет выводить результат номере полки
    """
    pasport = input('Введите номер документа: ') # сдесь вводится номер документа
    
    if shelf_pasport(pasport) != 'Документ не найден в базе':
        print('Документ хранится на полке:', shelf_pasport(pasport))
    else:
        print(shelf_pasport(pasport))
    
# ПУНКТ 3  
def all_info():
    """
    эта функция выводит все данные
    """
    for i in documents:
        print(f"№: {i['number']}, тип: {i['type']}, владелец: {i['name']}, полка хранения: {shelf_pasport(i['number'])}")
                                                             # в конце запрашивается полка передаваяв функцию номер паспорта
        
# ПУНКТ 4
def add_shelf():
    """
    эта функция добавляет полку
    """
    shelf = input('Введите номер полки: ')
    if shelf in directories.keys(): # если полка уже есть выведет что она есть
        a = list(directories.keys())
        shelfs = ', '.join(a) # соеденяю список
        print('Такая полка уже существует. Текущий перечень полок:', shelfs +'.')
    else:
        directories[shelf] = [] # добавит полку и пустой список для документов
        a = list(directories.keys())
        shelfs = ', '.join(a)
        print('Полка добавлена. Текущий перечень полок:', shelfs +'.')
    
    
# ПУНКТ 5
def del_shelf():
    """
    эта функция удаляет полку исходя из условий
    """
    shelf = input('Введите номер полки: ')
    
    if shelf in directories.keys() and directories[shelf] == []: # если полка есть и она пустая то удаляю
        del directories[shelf]
        a = list(directories.keys())
        shelfs = ', '.join(a)
        
        print('Полка удалена. Текущий перечень полок:', shelfs +'.')
        
    elif shelf in directories.keys() and directories[shelf] != []: # если полка есть но она не пустая то не удаляю
        a = list(directories.keys())
        shelfs = ', '.join(a)
        
        print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок:',shelfs +'.')
        
    else: # если полки нет выдаём шо её нет
        a = list(directories.keys())
        shelfs = ', '.join(a)
        
        print('Такой полки не существует. Текущий перечень полок:',shelfs +'.')
        
# ВСЕ ВЫЗОВЫ

action = 1
while action != 'q': # выполнять пока не введут это
    action = input('Введите команду: ')
    
    if action == 'p': # узнать владельца докумета
        main_pasport_name()       
    elif action == 's': # знать на какой полке храниться документ
        main_shelf_pasport()
    elif action =='j': # узнать всю инфу
        all_info()
    elif action =='as': # добавить полку
        add_shelf()
    elif action == 'ds': # удалить полку
        del_shelf()
    else:
        continue
```
