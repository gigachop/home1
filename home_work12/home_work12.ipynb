```python
for i in df:
    print(f'мода {i} - {df[i].mode()[0]}') # нахожу моду переберая названия столбцов
```

```python
# удоляю выбросы через межквартльный размах
q1 = df['Pulse'].quantile(0.25)
q3 = df['Pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(lower_bound, upper_bound)
remove_outliers = df[df['Pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Pulse')
remove_outliers
```

```python
# нахожу выбросы через межквартльный размах
q1 = df['Pulse'].quantile(0.25)
q3 = df['Pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(lower_bound, upper_bound)
outliers = df[(df['Pulse'] >= upper_bound) | (df['Pulse'] <= lower_bound)].sort_values('Pulse')
outliers
```

## Задание 3. Работа с пропусками

```python
df = pd.read_csv('horse_data.csv', header=None, na_values='?', names=columns, usecols=['Surgery','Age',
            'Rectal_Temperature','Pulse','Respiratory_Rate','Temperature_Of_Extremities', 'Pain','Outcome'])
df.head()
```

```python
pd.isnull(df).sum() # смотрю кол-во пропусков для столбцов
```

```python
df[df['Surgery'].isnull()] # в Surgery и Outcome всего 1 пропуско по этому можно удалить
```

```python
df = df.drop(index=[132], axis=1)
```

```python
# это не лучший вариант но другой не работает

df['Pulse'].fillna(df['Pulse'].mode()[0], inplace=True)
# в столбце пульса достаточно мало пропусков по этому можно заменить модой(самым распространённым значением)
```

```python
# df['Pulse'].fillna(df.groupby('Age')['Pulse'].transform('median'), inplace=True)
```

```python
# пульс вероятно зависит от возраста лошади, по этому лучше заменить модой(самым распространённым значением)
# сгрупировав по возрасту

df['Pulse'].fillna(df.groupby('Age')['Pulse'].transform('mode'), inplace=True)
```

```python
# в остальных столбцах много пропусков и по этому просто заменяю на no_info
df['Rectal_Temperature'].fillna('no_info', inplace = True)
df['Pain'].fillna('no_info', inplace = True)
df['Respiratory_Rate'].fillna('no_info', inplace = True)
df['Temperature_Of_Extremities'].fillna('no_info', inplace = True)
df['Pain'].fillna('no_info', inplace = True)
```
