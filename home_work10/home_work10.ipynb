### Задание 1

```python
df = pd.read_csv('visit_log.csv',sep=';')
df.head()
```

```python
def source_type(source):
    """
    эта функция берёт строчку и по  ней возвращяет source_type
    """
    
    if source['traffic_source'] == 'yandex' or source['traffic_source'] == 'google':
        return 'organic'
    elif (source['traffic_source'] == 'paid' or source['traffic_source'] == 'email') and source['region'] == 'Russia':
        return 'ad'
    elif (source['traffic_source'] == 'paid' or source['traffic_source'] == 'email') and source['region'] != 'Russia':
        return 'other'
    else:
        return source['traffic_source']
```

```python
df['source_type'] = df.apply(source_type,axis=1) # подаю всю строчку в функцию
df.head()
```

### Задание 3

```python
df = pd.read_csv('ratings.csv')
df.head()
```

```python
live = df.groupby('userId').agg(['min','max','count'])['timestamp'].reset_index()
# с помощью agg применяю несколько методов к стобцу timestamp, и групирую всё по пользователю
live.head()
```

```python
live_filtred = live[live['count'] > 100] # фильтрую пользователей у которых больше 100 оценок
live_filtred.head()
```

```python
live_filtred['livetime'] = live_filtred['max'] - live_filtred['min'] # нахожу время жизни
live_filtred.head()
```

```python
mean_live =live_filtred['livetime'].mean() # нахожу среднее время жизни среди всех пользователей
print(f'среднее время жизни пользователей, которые выставили более 100 оценок - {int(mean_live)}')
```
